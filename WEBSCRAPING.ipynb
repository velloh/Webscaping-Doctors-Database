```python
# Import webscraping libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the base URL
url = 'https://kmpdc.go.ke/Registers/practitioners.php'

# Initialize an empty list to store the data
all_data = []

# Before this step, inspect the website to get an idea of its structure
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html5lib')

# Feel free to glimpse the content at any/every stage
body = soup.find("body")
#print(body)

 # Extract table headers
headers = [header.text for header in body.find_all('th')]

 # Extract table rows
rows = body.find_all('tr')[1:]  # Skip the header row

 # loop through the rows, storing the cell data from each column into the empty list
for row in rows:
  cells = row.find_all('td')
  cell_data = [cell.text.strip() for cell in cells]
  all_data.append(cell_data)

# Create a DataFrame and append the now populated list
df = pd.DataFrame(all_data, columns=headers)

df

```

```python
from google.colab import files
# Save to CSV
df.to_csv('Registered_Medical_Practitioners_Kenya.csv', index=False)
files.download('Registered_Medical_Practitioners_Kenya.csv')
```
